```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\Dell\\Desktop\\Amazon_Sales_DataSet.csv")
print("First 5 records:")
display(df.head())
```

```python
print("Data Info:")
df.info()
```

```python
print("Missing Values:")
df.isnull().sum()
```

```python
print(df.dropna(subset=['product_name','discounted_price','actual_price','discount_percentage','rating']))
```

```python
top_categories = df['category'].value_counts().head(10)
top_categories.plot(kind='barh', title="Top 10 Product Categories", color='skyblue')
plt.xlabel('Number of Products')
plt.show()
```

```python
sns.heatmap(df[['discounted_price','actual_price','discount_percentage','rating']].corr(),annot=True,cmap='coolwarm')
plt.title('Correlation Between Price and Ratings')
plt.show()
```

```python
sns.histplot(df['rating'],bins=20,kde=True)
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Products")
plt.show()
```

```python
sns.scatterplot(data=df,x='discounted_price',y='rating',hue='category')
plt.title("Price vs Rating")
plt.show()
```
